```python
#Import calendar library to identify english names of month numbers easily
import calendar
```

```python
#Define readoriginaldate function
def readoriginaldate():
    #Ask user to enter a date given the prompt.
    date = input('Enter date in following format example: 6/28/12.\nDays and months can be 1 or 2 characters')
    #Print the inputted date.
    print('Date inputted:', date)
    #Return date variable.
    return date
```

```python
#Define breakoriginaldate date function. One parameter, the date, is entered.
def breakoriginaldate(date):
    #Prints how the date originally looked.
    print(date, 'is the original date.')
    #Split the date into a list where '/' is the separator.
    date = date.split('/')
    #Set the month, day, and year variables based on the index of the date list.
    month = date[0]
    day = date[1]
    year = date[2]
    #Print the month, day, and year in separate columns.
    print('Month:', month)
    print('Day:', day)
    print('Year:', year)
    #Return the month, day, and year variables.
    return month, day, year
```

```python
#Define printdate3ways function. 3 parameters are received: month, day, and year.
def printdate3ways(month, day, year):
    #Create europe variable and format the date in European style.
    europe = day+'-'+month+'-'+year
    #Print the European date style.
    print('European Printing:', europe)
    #Create english variable and format the date in English style.
    #Calender library used to transform month as a number to a word.
    #Since all dates are from 2000s, add 20 in front of the year.
    english = f'{calendar.month_name[int(month)]} {day}, 20{year}'
    #Print the English date style.
    print('English printing:', english)
    #Adds 0 in front of month or day if they are only one character
    if len(day) == 2:
        day = '0'+day
    if len(month) == 2:
        month = '0'+month
    #Create full variable and format the date in the full way style.
    full = day+'-'+month+'-'+'20'+year
    #Print the full date style.
    print('Full printing:', full)
    #Create space for next date
    print('\n\n')
```

```python
#Define the mainfunction function
def mainfunction():
    #While loop to ask user if they want to input a new date or if they are finished.
    while input("Type 'cancel' to end inputs") != 'cancel':
        #Bring back date variable from readoriginaldate function.
        #Runs readoriginaldate function
        date = readoriginaldate()
        #Bring back month, day, and year variables from breakoriginaldate function.
        #Runs breakoriginaldate function using date variable
        month, day, year = breakoriginaldate(date)
        #Runs printdate3ways function using month, day, and year variables.
        printdate3ways(month, day, year)
```

```python
#Run mainfunction function
mainfunction()
```
