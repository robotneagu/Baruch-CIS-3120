```python
#Introduction function. 
#No parameters received. Prints the instructions of what the program does.

def introduction():
    print('Comparing 3 Numbers to the Average')
    print('by Robert Neagu\n')

    print('You will be prompted to enter 3 individual integers.')
    print('Then, the program will return the average and \nwhether each integer was above, below, or equal to the average.')
    print('A message will also state how many values are equal to the average.\n')

    print('Please do not enter a non-integer value at any input!\nDoing so will break the code as it only accepts integers as an input.\n')

    print('How to end the program:')
    print('Type the value 0 for all 3 prompts.')
    print('The same procedure will follow, but you will see')
    print('how many sets of values you entered.\n')
```

```python
#Average function
#Receives the list of 3 numbers, list, and calculates the average. #Only returns the average.

def findaverage(list):
    avg_calc = sum(list)/len(list)
    return avg_calc
```

```python
#Compare to average function
#Receives the number list, list, and the average, avg.
#Takes each number in the list and compares it to the average.
#Will print whether each number is above, equal to, or below the average.

def comparetoavg(list, avg):
    equal_avg = 0

    for i in range (len(list)):
        current_num = int(list[i])
        if current_num > avg:
            print(f'{current_num} is above the average.')
        elif current_num == avg:
            print(f'{current_num} is equal to the average.')
            equal_avg += 1
        else:
            print(f'{current_num} is less than the average')

    if equal_avg == 1:
        print(f'{equal_avg} value is equal to the average.')
    else:
        print(f'{equal_avg} values are equal to the average.')
```

```python
#Main function
#No parameters received. Creates the number list to edit and counts how many times the list was edited.
#Executes all above functions.
#Generates the special combination, "0, 0, 0", that will end the program.

def mainfunction():
    introduction()
    count_set = 0
    number_list = [1, 1, 1]

    while number_list != [0, 0, 0]:

        for i in range(len(number_list)):
            int_input = int(input(f'Enter integer #{i+1}:'))
            number_list[i] = int_input
        print(number_list)
        
        #Round average to 3 decimal places
        avg_calc = findaverage(number_list)
        avg_calc = round(avg_calc, 3)
        print(avg_calc)

        comparetoavg(number_list, avg_calc)

        count_set += 1

        for i in range(4):
            print('')
    
    print('Number of sets of values entered:')
    print(count_set)
```

```python
mainfunction()
```
